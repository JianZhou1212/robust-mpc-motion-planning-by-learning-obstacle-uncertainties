```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
from scipy.io import savemat
from shapely.geometry import Polygon, Point
```

```python
%matplotlib inline
```

## Define Parameters and Functions for Visualization

```python
RefPos_EV =  np.load('RefPos_EV.npy', allow_pickle=True)
RefPos_SV =  np.load('RefPos_SV.npy', allow_pickle=True)
l_veh_sv = 0.36 # vehicle length
w_veh_sv = 0.23 # vehicle width
l_veh_ev = 0.26 # vehicle length
w_veh_ev = 0.25 # vehicle width
x_ref_EV = RefPos_EV[0] # EV reference x
y_ref_EV = RefPos_EV[1] # EV reference y
v_ref_EV = 0            # EV reference v
phi_ref_EV = 0          # EV reference phi
x_low  = 0 # min x position, m
y_low  = 0 # min y position, m
x_up   = 8 # max x position, m
y_up   = 8 # max y position, m
Road  = Polygon([(x_low, y_low), (x_up, y_low), (x_up, y_up), (x_low, y_up)]) # driveable area, subtract the vehicle size

def rota_rect(state, l, w):
    # Visualize the vehicle shape based on pos. and heading
    # Note: the first four states must be x, y, heading, velocity
    
    x       = state[0]
    y       = state[1]
    box     = [(x - l/2, y + w/2), (x + l/2, y + w/2), (x + l/2, y - w/2), (x - l/2, y - w/2)]
    heading = state[2]
    
    box_matrix = np.array(box) - np.repeat(np.array([[x, y]]), len(box), 0)
    theta = -heading
    rota_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

    new = box_matrix.dot(rota_matrix) + np.repeat(np.array([[x, y]]), len(box), 0)
    
    x = [new[0][0], new[1][0], new[2][0], new[3][0], new[0][0]]
    y = [new[0][1], new[1][1], new[2][1], new[3][1], new[0][1]]
    return x, y

def ploy_distance(x_pos_1, y_pos_1, x_pos_2, y_pos_2):
    # Compute the distance between two obejects
    # If the output is 0, means collision
    obj1 = Polygon([(x_pos_1[0],y_pos_1[0]), (x_pos_1[1],y_pos_1[1]), (x_pos_1[2],y_pos_1[2]), (x_pos_1[3],y_pos_1[3])])
    obj2 = Polygon([(x_pos_2[0],y_pos_2[0]), (x_pos_2[1],y_pos_2[1]), (x_pos_2[2],y_pos_2[2]), (x_pos_2[3],y_pos_2[3])])
    
    return obj1.distance(obj2)

def road_collision(x, y, Road):
    # Chech if the object is in the drivable area
    # If the output is 0, in the area
    return np.max(np.array([Road.distance(Point(x[0], y[0])), Road.distance(Point(x[1], y[1])), Road.distance(Point(x[2], y[2])), Road.distance(Point(x[3], y[3]))]))
```

## Load Data

```python
t = np.load('t.npy')

State_SV   = np.load('State_SV.npy')
Control_SV = np.load('Control_SV.npy', allow_pickle=True)

State_EV_D = np.load('State_EV_D.npy')
State_EV_P = np.load('State_EV_P.npy')
State_EV_R = np.load('State_EV_R.npy')

Control_EV_D = np.load('Control_EV_D.npy')
Control_EV_P = np.load('Control_EV_P.npy')
Control_EV_R = np.load('Control_EV_R.npy')

J_D = np.load('J_D.npy', allow_pickle=True)
J_P = np.load('J_P.npy', allow_pickle=True)
J_R = np.load('J_R.npy', allow_pickle=True)

X_SV_Planning    = np.load('X_SV_Planning.npy', allow_pickle = True)
X_EV_Planning_D  = np.load('X_EV_Planning_D.npy', allow_pickle = True)
X_EV_Planning_P  = np.load('X_EV_Planning_P.npy', allow_pickle = True)
X_EV_Planning_R  = np.load('X_EV_Planning_R.npy', allow_pickle = True)
```

## Compute the Minimal Distance Between EV and SV

```python
Min_Dis_D_SV   = list( ) # Distacne with SV, Deterministic
Min_Dis_D_Road = list( ) # Distance with road boundary, Deterministic (if 0, EV is in the driable area)
Min_Dis_P_SV   = list( ) # Distacne with SV, Proposed
Min_Dis_P_Road = list( ) # Distance with road boundary, Proposed (if 0, EV is in the driable area)
Min_Dis_R_SV   = list( ) # Distacne with SV, Robust
Min_Dis_R_Road = list( ) # Distance with road boundary, Robust (if 0, EV is in the driable area)

for k in range(len(t)):

    x_SV, y_SV    = rota_rect(State_SV[:, k], l_veh_sv, w_veh_sv)
    x_EV_D, y_EV_D = rota_rect(State_EV_D[:, k], l_veh_sv, w_veh_sv)
    x_EV_P, y_EV_P = rota_rect(State_EV_P[:, k], l_veh_ev, w_veh_ev)
    x_EV_R, y_EV_R = rota_rect(State_EV_R[:, k], l_veh_ev, w_veh_ev)

    Min_Dis_D_SV.append(ploy_distance(x_SV, y_SV, x_EV_D, y_EV_D))
    Min_Dis_P_SV.append(ploy_distance(x_SV, y_SV, x_EV_P, y_EV_P))
    Min_Dis_R_SV.append(ploy_distance(x_SV, y_SV, x_EV_R, y_EV_R))
    
    Min_Dis_D_Road.append(road_collision(x_EV_D, y_EV_D, Road))
    Min_Dis_P_Road.append(road_collision(x_EV_P, y_EV_P, Road))
    Min_Dis_R_Road.append(road_collision(x_EV_R, y_EV_R, Road))
```

## Compute the Distance to the Reference

```python
Distance_to_ref_D = np.sqrt((State_EV_D[0, :] - x_ref_EV)**2 + (State_EV_D[1, :] - y_ref_EV)**2) +(State_EV_D[2, :] - phi_ref_EV)**2 + (State_EV_D[3, :] - v_ref_EV)**2
Distance_to_ref_P = np.sqrt((State_EV_P[0, :] - x_ref_EV)**2 + (State_EV_P[1, :] - y_ref_EV)**2) +(State_EV_P[2, :] - phi_ref_EV)**2 + (State_EV_P[3, :] - v_ref_EV)**2
Distance_to_ref_R = np.sqrt((State_EV_R[0, :] - x_ref_EV)**2 + (State_EV_R[1, :] - y_ref_EV)**2) +(State_EV_R[2, :] - phi_ref_EV)**2 + (State_EV_R[3, :] - v_ref_EV)**2
```

## Show the Global Path

```python
# Deterministic
plt.figure(figsize=(4, 4), tight_layout = True)
for k in range(len(t)):
    
    x_SV, y_SV = rota_rect(State_SV[:, k], l_veh_sv, w_veh_sv)
    x_EV_D, y_EV_D = rota_rect(State_EV_D[:, k], l_veh_sv, w_veh_sv)

    plt.plot(x_SV, y_SV, c = 'r', linewidth = 1)
    plt.plot(x_EV_D, y_EV_D, c = 'b', linewidth = 1)

plt.title("Deterministic")   
plt.xlabel("x [m]")
plt.ylabel("y [m]")

# Proposed
plt.figure(figsize=(4, 4), tight_layout = True)
for k in range(len(t)):
    
    x_SV, y_SV = rota_rect(State_SV[:, k], l_veh_sv, w_veh_sv)
    x_EV_P, y_EV_P = rota_rect(State_EV_P[:, k], l_veh_ev, w_veh_ev)

    plt.plot(x_SV, y_SV, c = 'r', linewidth = 1)
    plt.plot(x_EV_P, y_EV_P, c = 'b', linewidth = 1)

plt.title("Proposed")    
plt.xlabel("x [m]")
plt.ylabel("y [m]")

# Robust
plt.figure(figsize=(4, 4), tight_layout = True)
for k in range(len(t)):
    
    x_SV, y_SV = rota_rect(State_SV[:, k], l_veh_sv, w_veh_sv)
    x_EV_R, y_EV_R = rota_rect(State_EV_R[:, k], l_veh_ev, w_veh_ev)
    
    plt.plot(x_SV, y_SV, c = 'r', linewidth = 1)
    plt.plot(x_EV_R, y_EV_R, c = 'b', linewidth = 1)

plt.title("Robust")       
plt.xlabel("x [m]")
plt.ylabel("y [m]")
```
